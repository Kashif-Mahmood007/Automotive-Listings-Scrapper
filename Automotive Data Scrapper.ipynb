```python
## As We have a data of Multiple pages, so we need to run loop and store data in lists
```

```python
# Creating Lists for storing all Data

Name = []
Price = []
Location = []
Year = []
Mileage = []
Fuel_Type = []
Engine_Capacity = []
Transmission = []
```

```python
## Get the main div which includes all the stuff
```

```python
cars = soup.find_all('li', class_= 'classified-listing')
len(cars)
```

```python
for i in range(1, 1000):
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(f".....?page={i}", headers = headers).text
    soup = BeautifulSoup(webpage, 'lxml')
    cars = soup.find_all('li', class_= 'classified-listing')

    for i in cars:
        try:
            Name.append(i.find('h3').text.strip())
        except:
            Name.append(np.nan)

        try:
            Price.append(i.find('div', class_ = 'price-details').text.strip())
        except:
            Price.append(np.nan)

        try:
            Location.append(i.find('ul', class_ = 'search-vehicle-info').text.strip())
        except:
            Location.append(np.nan)

        ul = i.find('ul', class_= 'search-vehicle-info-2')

        try:
            Year.append(ul.find_all('li')[0].text.strip())
        except:
            Year.append(np.nan)

        try:
            Mileage.append(ul.find_all('li')[1].text.strip())
        except:
            Mileage.append(np.nan)

        try:
            Fuel_Type.append(ul.find_all('li')[2].text.strip())
        except:
            Fuel_Type.append(np.nan)

        try:
            Engine_Capacity.append(ul.find_all('li')[3].text.strip())
        except:
            Engine_Capacity.append(np.nan)

        try:
            Transmission.append(ul.find_all('li')[4].text.strip())
        except:
            Transmission.append(np.nan)

    
```

```python
## Now, It's time to make a DataFrame
```

```python
df = pd.DataFrame({
    'Name' : Name,
    'Price' : Price,
    'Location' : Location,
    'Year' : Year,
    'Mileage' : Mileage,
    'Fuel_Type' : Fuel_Type,
    'Engine_Capacity' : Engine_Capacity,
    'Transmission' : Transmission
})
```
